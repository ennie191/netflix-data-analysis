```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/data.csv").head(10000)
df.dropna(subset=["title", "genres", "releaseYear", "imdbAverageRating"], inplace=True)

# Initial preview
print(df.head())
print(df.info())
print(df.describe())

# Check for missing values
missing = df.isnull().sum()
print("Missing values:\n", missing)

# Genre preprocessing
df["genres"] = df["genres"].apply(lambda x: x.split(", "))

# Most frequent genres
from collections import Counter
all_genres = [genre for sublist in df["genres"] for genre in sublist]
common_genres = Counter(all_genres).most_common(10)
print("Top Genres:", common_genres)

# Ratings distribution
sns.histplot(df["imdbAverageRating"], kde=True)
plt.title("IMDb Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

# Release year distribution
sns.histplot(df["releaseYear"], bins=30)
plt.title("Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
